## Mapping 2020 precicnt election results by state house district

```python
# Import relevant packages

import geopandas as gpd
import pandas as pd
import folium
from matplotlib import pyplot as plt
import shapefile
import matplotlib.patheffects as pe
import adjustText as aT
from adjustText import adjust_text
```

```python
# Read in shapefiles, filter data

shp_df = gpd.read_file('Precincts.shp')
filtered_shp = shp_df[(shp_df.CNTY == 201)]
filtered_shp = filtered_shp[['PREC', 'geometry']]

# hd_shp_df = gpd.read_file('C:/Users/Nathan/Documents/GIS/PLANH358/PLANH358.shp')
# filtered_hd_shp = hd_shp_df[(hd_shp_df.District >= 126) & (hd_shp_df.District != 136)]
```

```python
# Read in data, specify column types, and usecols

col_types = {
    'Precinct': object,
    'State Rep': 'category',
    'Registered Voters': int,
    'Total Votes': int,
    'Voter Turnout': float,
    'Trump Net': int,
    'State Rep Net': int,
    'HD GOP': int,
    'HD DEM': int,
    'Cornyn': int,
    'Hegar': int,
    'Cornyn Net': int
}
use_cols = ['Precinct', 'State Rep', 'Cornyn Net', 'Trump Net', 'State Rep Net']
filename = 'C:/Users/Nathan/Documents/2020/Election Analysis/2020 General Election Analysis/Murphy Folder/2020 Landscape.csv'
df = pd.read_csv(filename, dtype=col_types, usecols = use_cols)
```

```python
# Reformat Precincts to include leading zeros to match

df['Precinct'] = df['Precinct'].str.zfill(4)
```

```python
# Merge dataframes

merge_df = filtered_shp.merge(df, how='left', left_on='PREC', right_on='Precinct')
```

```python
# Create center points for polygons for labeling

merge_df['center'] = merge_df['geometry'].centroid
```

```python
# Define function to map districts

def MapDistrict(district, candidate='State Rep'):
    
    # Create title of plot
    title = candidate+' Net Votes: '+str(merge_df.loc[merge_df['State Rep'] == 'SR '+district, str(candidate+' Net')].sum())

    # Create temporary df of filtered district
    temp_df = merge_df[merge_df['State Rep'] == ('SR '+district)]
    
    # Create df of center points for labeling
    merge_df_points = temp_df.copy()
    merge_df_points.set_geometry('center', inplace = True)

    # Create figure and axis for plot, plot geometry
    fig, ax1 = plt.subplots(figsize=(20,10))
    temp_df.plot(str(candidate+' Net'), cmap='seismic', edgecolor='lightgrey', linewidth=1.5, legend=True, ax=ax1, vmin=-2000, vmax=2000)

    # Create labels for polygons (precincts)
    texts = []
    for x, y, label in zip(merge_df_points.geometry.x, merge_df_points.geometry.y, merge_df_points['PREC']):
        texts.append(plt.text((x-300), y, label, fontsize = 11, fontname='lato', path_effects=[pe.withStroke(linewidth=4, foreground="white")]))

    aT.adjust_text(texts, force_points=0.3, force_text=0.8, expand_points=(2,1), expand_text=(1,1), 
                    arrowprops=dict(arrowstyle="-", color='#414a4c', lw=0.5))

    adjust_text(texts)

    # Format title, axis
    plt.title(title, loc='left',fontsize=24, fontname='lato')
    ax1.set_axis_off()
    
    # Save map image to directory
    filename = 'HD' + str(district) + '_' + candidate + '_Map.png'
    plt.savefig(filename, bbox_inches = 'tight', pad_inches = 0.25, dpi=150)
```

```python
# Test function

MapDistrict('135')
```
